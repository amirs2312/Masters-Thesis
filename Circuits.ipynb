```python
from qiskit import *
from qiskit import QuantumCircuit
from qiskit import transpile
from qiskit.circuit import Parameter
from qiskit.circuit.library import RYGate
from qiskit.circuit import Parameter
import pdflatex
import pylatexenc
import numpy as np
```

# Linear Entangling Circuit

```python
linear_vqc = QuantumCircuit(4)

theta = Parameter('')
phi = Parameter('')

for i in range(4):
    linear_vqc.ry(phi, i)

linear_vqc.barrier()


for i in range(4):
    linear_vqc.h(i)

for i in range(3):
    linear_vqc.cx(i, i + 1)


linear_vqc.barrier()


for i in range(4):
    linear_vqc.ry(theta, i)



linear_vqc.draw(output="latex", scale=2)
```

# Cyclic Entangling Circuit

```python
cyclic_vqc = QuantumCircuit(4)

theta = Parameter('')
phi = Parameter('')

for i in range(4):
    cyclic_vqc.ry(phi, i)

cyclic_vqc.barrier()
for i in range(4):
    cyclic_vqc.h(i)

for i in range(3):
    cyclic_vqc.cx(i, i + 1)
cyclic_vqc.cx(3, 0)

cyclic_vqc.barrier()

for i in range(4):
    cyclic_vqc.ry(theta, i)



cyclic_vqc.draw(output="latex", scale=2)
```

# Star Entangling Circuit

```python
star_vqc = QuantumCircuit(4)

theta = Parameter('')
phi = Parameter('')

for i in range(4):
    star_vqc.ry(phi, i)

star_vqc.barrier()

for i in range(4):
    star_vqc.h(i)

for i in range(4):
    if i != 2:
        star_vqc.cx(i, 2)


star_vqc.barrier()

for i in range(4):
    star_vqc.ry(theta, i)



star_vqc.draw(output="latex", scale=2)
```

# Fully Entangling Circuit

```python
full_vqc = QuantumCircuit(4)

theta = Parameter('')
phi = Parameter('')

for i in range(4):
    full_vqc.ry(phi, i)

full_vqc.barrier()

for i in range(4):
    full_vqc.h(i)

for i in range(4):
    for j in range(4):
        if i != j:
            full_vqc.cx(i, j)


full_vqc.barrier()

for i in range(4):
    full_vqc.ry(theta, i)



full_vqc.draw(output="latex", scale=2)
```
